```python

import pandas as pd
import matplotlib.pyplot as plt
##compare all test result 
dataset = 'sz'
pre_len = 1
epoch = 100
path = f"out/tgcn/tgcn_{dataset}_lr0.001_batch32_unit64_seq12_pre{pre_len}_epoch{epoch}_adj-dist/test_result.xlsx"
data = pd.read_excel(path, sheet_name='pred', header=None)#, engine='openpyxl')
test_result = data.values
data = pd.read_excel(path, sheet_name='true', header=None)#, engine='openpyxl')
true_labels = data.values

path = f'out/tgcn/tgcn_{dataset}_lr0.001_batch32_unit64_seq12_pre{pre_len}_epoch{epoch}_adj-gsl/test_result.xlsx'
data = pd.read_excel(path, sheet_name='pred', header=None)#, engine='openpyxl')
test_result_gsl = data.values
# data = pd.read_excel(path, sheet_name='true', header=None)
# true_labels = data.values
path = f'out/tgcn/tgcn_{dataset}_lr0.001_batch32_unit64_seq12_pre{pre_len}_epoch{epoch}_adj-gsldist/test_result.xlsx'
data = pd.read_excel(path, sheet_name='pred', header=None)#, engine='openpyxl')
test_result_gsldist = data.values

path = 'out/tgcn'
fig1 = plt.figure(figsize=(15,5))
#    ax1 = fig1.add_subplot(1,1,1)
a_pred = test_result[:,0]
a_pred_gsl = test_result_gsl[:,0]
a_pred_gsldist = test_result_gsldist[:,0]
a_true = true_labels[:,0]

TGCN_w_adj_caption = 'T-GCN (w/ adj)'
TGCN_wo_adj_caption = 'T-GCN (w/o adj)'
TGCN_w_adj_GSL_caption = TGCN_w_adj_caption+'+GSL'
TGCN_wo_adj_GSL_caption = TGCN_wo_adj_caption+'+GSL'

fig1 = plt.figure(figsize=(17,5))
plt.plot(a_true,'b-',label='Ground Truth', linewidth=1.0)
plt.plot(a_pred,'r--',label=TGCN_w_adj_caption, linewidth=2.0)
plt.plot(a_pred_gsl,'g-',label=TGCN_wo_adj_GSL_caption, linewidth=3.0)
plt.plot(a_pred_gsldist,'k:',label=TGCN_w_adj_GSL_caption, linewidth=2.0)
plt.legend(loc='best',fontsize=15)
file_name = f'/tgcn_{dataset}_lr0.001_batch32_unit64_seq12_pre{pre_len}_epoch{epoch}_adj_compare_all.jpg'
plt.savefig(path+file_name)
plt.show()
## oneday test result visualization
# fig1 = plt.figure(figsize=(17,5))
#    ax1 = fig1.add_subplot(1,1,1)
# First test day
a_pred = test_result[:96,0]
a_pred_gsl = test_result_gsl[:96,0]
a_pred_gsldist = test_result_gsldist[:96,0]
a_true = true_labels[0:96,0]

fig1 = plt.figure(figsize=(17,5))
plt.plot(a_true,'b-',label='Ground Truth', linewidth=1.0)
plt.plot(a_pred,'r--',label=TGCN_w_adj_caption, linewidth=2.0)
plt.plot(a_pred_gsl,'g-',label=TGCN_wo_adj_GSL_caption, linewidth=3.0)
plt.plot(a_pred_gsldist,'k:',label=TGCN_w_adj_GSL_caption, linewidth=2.0)
plt.legend(loc='best',fontsize=15)
file_name = f'/tgcn_{dataset}_lr0.001_batch32_unit64_seq12_pre{pre_len}_epoch{epoch}_adj_test_firstday.jpg'
plt.savefig(path+file_name)
plt.show()

# Last test day
a_pred = test_result[-96:,0]
a_pred_gsl = test_result_gsl[-96:,0]
a_pred_gsldist = test_result_gsldist[-96:,0]
a_true = true_labels[-96:,0]

fig1 = plt.figure(figsize=(17,5))
plt.plot(a_true,'b-',label='Ground Truth', linewidth=1.0)
plt.plot(a_pred,'r--',label=TGCN_w_adj_caption, linewidth=2.0)
plt.plot(a_pred_gsl,'g-',label=TGCN_wo_adj_GSL_caption, linewidth=3.0)
plt.plot(a_pred_gsldist,'k:',label=TGCN_w_adj_GSL_caption, linewidth=2.0)
plt.legend(loc='best',fontsize=15)
file_name = f'/tgcn_{dataset}_lr0.001_batch32_unit64_seq12_pre{pre_len}_epoch{epoch}_adj_test_lastday.jpg'
plt.savefig(path+file_name)
plt.show()
```
